```python
import os

bashrc_path = os.path.expanduser("~/.bashrc")
with open(bashrc_path, "r") as f:
    for line in f:
        if line.strip().startswith("export "):
            key, _, val = line.strip().replace("export ", "").partition("=")
            if key != "PATH":
                os.environ[key] = val.strip().strip('"')

bucket_input   = os.environ.get("WORKSPACE_BUCKET")
bucket_output  = os.environ.get("WORKSPACE_BUCKET_OUTPUT")
bucket_scripts = os.environ.get("WORKSPACE_BUCKET_SCRIPTS")
bucket_GenDat  = os.environ.get("WORKSPACE_BUCKET_GENDAT")
dataset        = os.environ.get("WORKSPACE_CDR")

print(f"bucket_input:   {bucket_input}")
print(f"bucket_output:  {bucket_output}")
print(f"bucket_scripts: {bucket_scripts}")
print(f"bucket_GenDat: {bucket_GenDat}")
print(f"dataset:        {dataset}")
```
